```python
doc = nlp(df['Resume_str'][0])
```

```python
# remove stopwords
stopwords = nlp.Defaults.stop_words

text = doc.text

lst=[]
for token in text.split():
    if token.lower() not in stopwords:    #checking whether the word is not 
        lst.append(token)                    #present in the stopword list.

doc = nlp(' '.join(lst).lower())    
```

```python
chunks = set()
individual_words = set()
for chunk in doc.noun_chunks:
#     print(chunk.text, chunk.root.text, chunk.root.dep_,
#             chunk.root.head.text)
#     print(chunk.root.head.text)
    chunks.add(chunk.text)
    individual_words.add(chunk.root.head.text)
individual_words = list(individual_words)
    
clean_chunks = []  
for chunk in chunks: 
#     chunks2.union(set(chunk.split(", ")))
    clean_chunks.extend(re.split(r"(, | － )", chunk))
```

```python
filename = "./linkedinskill"
f = open(filename, 'rb')
Lines = f.readlines()
skills = []
for line in Lines:
    line = line[:-1].decode("utf-8")
    skills.append(line[:-1].lower())
skills
```

```python
#define binary search
def bs(target, arr):
    lo = 0
    hi = len(arr)-1
    while lo <= hi:
        mid = (lo + hi)//2
        if arr[mid] == target:
            return True
        elif target < arr[mid]:
            hi = mid-1
        else:
            lo = mid+1
    return False 
```
